```python
#setting the data type of selected columns
#stringing the COUNTRY, column 6
Data1['COUNTRY']=Data1['COUNTRY'].apply(lambda x: str(x))
#stringing the COUNTRY, column 7
Data1['LATITUDE']=Data1['LATITUDE'].apply(lambda x: str(x))
#stringing the COUNTRY, column 28
Data1['WEATHER.CONDITION']=Data1['WEATHER.CONDITION'].apply(lambda x: str(x))
#setting the PUBLIC.DATE date datatype
Data1['PUBLICATION.DATE']=Data1['PUBLICATION.DATE'].apply(lambda x: str(x))
```

```python
#checking null values
Data1.isna().sum()
```

```python
#dropping the null values
Data2=Data1.dropna(how='all')
Data2
```

```python
Data2.fillna(0, inplace=True)
Data2
```

```python
Data2.dtypes
```

```python
#checking on the duplicates
Data2.duplicated().any()
```

```python
#creating the column containg injuries summation
Data2["TOTAL_INJURIES_SUMMATION"]=Data2["TOTAL.FATAL.INJURIES"]+Data2["TOTAL.MINOR.INJURIES"]+Data2["TOTAL.UNINJURED"]

```

```python
#looking for the most dangerouse plane
DANGEROUSE_PLANES=Data2["TOTAL_INJURIES_SUMMATION"].max(axis=0)
DANGEROUSE_PLANES
```

```python
#abstracting the dangerouse plane from the dataset
DANGEROUSE_PLANE=Data2[(Data2["TOTAL_INJURIES_SUMMATION"]==576.0)]
DANGEROUSE_PLANE
```

```python
#looking for favourable and safe planes 
BETTER_AND_SAFE_PLANES1=Data2["TOTAL_INJURIES_SUMMATION"].min(axis=0)
BETTER_AND_SAFE_PLANES1
```

```python
#abstracting the safer and better plane from the dataset
BETTER_AND_SAFE_PLANES=Data2[(Data2["TOTAL_INJURIES_SUMMATION"]==0.0)]
BETTER_AND_SAFE_PLANES
```

```python
#obtaining the aircrafts that had minor damages from the BETTER_AND_SAFE_PLANES
SUBSTANTIAL_AIRCRAFTS=BETTER_AND_SAFE_PLANES[(BETTER_AND_SAFE_PLANES["AIRCRAFT.DAMAGE"]=="Minor")]
SUBSTANTIAL_AIRCRAFTS
```

```python
#Obtaining aircrafts with insident injuries only
AIRCRAFT_WITH_INSIDENT_INJURIES=SUBSTANTIAL_AIRCRAFTS[(SUBSTANTIAL_AIRCRAFTS["INJURY.SEVERITY"]=="Incident")]
AIRCRAFT_WITH_INSIDENT_INJURIES
```

```python
#TRYING TO OBTAIN THE BUSINESS aircrafts fro above AIRCRAFT_WITH_INSIDENT_INJURIES
BUSINESS_AIECRAFTS=AIRCRAFT_WITH_INSIDENT_INJURIES[(AIRCRAFT_WITH_INSIDENT_INJURIES["PURPOSE.OF.FLIGHT"]=="Business")]
BUSINESS_AIECRAFTS
```
